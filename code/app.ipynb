```python

url = 'https://walmart.wd5.myworkdayjobs.com/en-US/WalmartExternal?clientRequestID=5af893c0f55c4f7d9c4ff14336a46c1b&jobFamilyGroup=e83ebdbd2a0a01af0185848948e94dc6&locationCountry=bc33aa3152ec42d4995f4791a106ed09'
# Setup Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode
chrome_options.add_argument("--no-sandbox") # Bypass OS security model
chrome_options.add_argument("--disable-dev-shm-usage") # overcome limited resource problems

driver = webdriver.Chrome(options=chrome_options)

driver.get(url)
sleep(10)
content = driver.page_source
soup = BeautifulSoup(content, 'html.parser') # Specify the parser explicitly
```

```python
# soup.find_all('body')[0].find_all('div')[0].find_all('div')[0]\
#   .find_all('div')[0].find_all("ul")[1].find_all('li')
```

```python
li_tags = soup.find_all('li', class_='css-1q2dra3')
```

```python
data = []
for li in li_tags:
    title = li.find('a', {'data-automation-id': 'jobTitle'}).get_text(strip=True)
    loc = li.find('dd', {'class': 'css-129m7dg'}).get_text(strip=True)
    posted_on = li.find('div', { 'data-automation-id': 'postedOn'}).find('dd').get_text(strip=True)
    uri = li.find('a', {'data-automation-id': 'jobTitle'}).get('href')

    # Append to the data list
    data.append({
        'title': title,
        'loc': loc,
        'posted on': posted_on,
        'URI': uri,
        'company': 'Walmart',
        'source': 'Workday',
        'type': 'FT',
        'extracted_at': datetime.now(pytz.timezone('US/Central')).strftime("%Y-%m-%d %H:%M:%S")

    })
```

```python
df = pd.DataFrame(data)

df['URI']= df["URI"].apply(lambda x: f"https://walmart.wd5.myworkdayjobs.com{x}")

# Print DataFrame
(df)
```

```python


# Assuming 'df' is your pandas DataFrame
# Replace 'your_collection_name' with your desired Firestore collection name

def dataframe_to_firestore(df, collection_name):
    for index, row in df.iterrows():
        try:
            doc_ref = db.collection(collection_name).document()
            doc_ref.set(row.to_dict())
        except Exception as e:
            print(f"Error adding document {index}: {e}")


# Example Usage with the two DataFrames:
# dataframe_to_firestore(df)
dataframe_to_firestore(cvs_rdf, 'test_cvs_jobs')
```
